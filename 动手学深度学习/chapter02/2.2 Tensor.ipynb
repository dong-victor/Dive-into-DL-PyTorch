```python
import matplotlib.pyplot as plt
# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# 生成100000个样本
rand_samples = torch.rand(100000)
randn_samples = torch.randn(100000)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))

ax1.hist(rand_samples.numpy(), bins=50, density=True, alpha=0.7, color='blue')
ax1.set_title('torch.rand(均匀分布)')
ax1.set_xlabel('值')
ax1.set_ylabel('概率密度')

ax2.hist(randn_samples.numpy(), bins=50, density=True, alpha=0.7, color='green')
ax2.set_title('torch.randn(标准正态分布)')
ax2.set_xlabel('值')
ax2.set_ylabel('概率密度')

plt.tight_layout()
plt.show()
```
