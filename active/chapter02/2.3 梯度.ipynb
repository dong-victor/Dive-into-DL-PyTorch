```python
import torch
```

```python
x = torch.ones(2, 2, requires_grad=True)
print(x)
print(x.grad_fn)
```

```python
y = x + 2
print(y)
print(y.grad_fn)
```

```python
print(x.is_leaf, y.is_leaf)
```

```python
z = y * y * 3
out = z.mean()
print(z, out)
```

```python
# 创建需要梯度的张量
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
print(f"张量 x: {x}")
print(f"requires_grad: {x.requires_grad}")

# 普通张量（不需要梯度）
y = torch.tensor([1.0, 2.0, 3.0], requires_grad=False)
print(f"普通张量: {y.requires_grad}")
```

```python
# 简单例子
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2 + 3 * x + 1

# 反向传播计算梯度
y.backward()

# 查看梯度 dy/dx = 2x + 3 = 2*2 + 3 = 7
print(f"x 的梯度: {x.grad}")  # tensor(7.)
```

```python
x = torch.tensor(3.0, requires_grad=True)
y = x ** 3 + x ** 3 + 7 * x + 12

y.backward()
print(x.grad)
```
