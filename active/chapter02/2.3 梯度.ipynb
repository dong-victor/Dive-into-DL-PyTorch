```python
import torch
```

```python
x = torch.ones(2, 2, requires_grad=True)
print(x)
print(x.grad_fn)
```

```python
y = x + 2
print(y)
print(y.grad_fn)
print(x.grad_fn)
```

```python
print(x.is_leaf, y.is_leaf)
```

```python
z = y*y*3
out = z.mean()
print(z, out)
print(z.grad_fn, out.grad_fn)
```

```python
out.backward()
print(x.grad)
```

```python
out2 = x.sum()
out2.backward()
print(x.grad)

out3 = x.sum()
x.grad.data.zero_()
out3.backward()
print(x.grad)
```

```python
x = torch.tensor([1.0, 2.0, 3.0, 4.0], requires_grad=True)
y = 2 * x
z = y.view(2,2)
print(z)
```

```python
v = torch.tensor([[1.0, 0.1], [0.01, 0.001]], dtype=torch.float)
z.backward(v)
print(x.grad)
```
