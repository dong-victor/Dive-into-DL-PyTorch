## 下载一个数据集

```python
import torchvision
from torchvision import transforms
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter('../logs')
```

## 使用transforms转换图片

```python
data_transforms = transforms.Compose([
    transforms.ToTensor()
])

data_train = torchvision.datasets.CIFAR10('../dataset', train=True, transform= data_transforms, download=True)
data_test = torchvision.datasets.CIFAR10('../dataset', train=False, transform= data_transforms, download=True)
```

## 使用 tensorBoard显示

```python
for i in range(10):
    image, label = data_train[i]
    print(type(image), data_train.classes[label])
    writer.add_image('CIFAR10', image, i)
```

```python
writer.close()
```
