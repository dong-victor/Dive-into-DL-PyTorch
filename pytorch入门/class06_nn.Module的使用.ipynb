```python
import torch.nn as nn
import torch
from PIL import Image
from torchvision.transforms import transforms
import torch.nn.functional as F
```

```python
class TestConv(nn.Module):
    def __init__(self):
        super(TestConv, self).__init__()
        self.conv1 = nn.Conv2d(3, 5, 3)
        self.conv2 = nn.Conv2d(5, 1, 3)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        return F.relu(self.conv2(x))
```

```python
x = torch.randn(3 * 5 * 5).reshape((3, 5, 5))
x
```

```python
test = TestConv()
y = test(x)
print(y)
```

```python
image = Image.open('../dataset/0Q3A0032.JPG')

image_compose = transforms.Compose([
    transforms.Resize(32),
    transforms.ToTensor()
])
image_r = image_compose(image)
print(type(image_r))
y = test(image_r)
print(y)
```
