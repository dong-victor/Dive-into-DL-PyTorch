## nn.Conv2

```python
from torch import nn
from torch.utils.data import DataLoader
import torch
import torchvision
from torch.utils.tensorboard import SummaryWriter
from torchvision import transforms

writer = SummaryWriter('../logs')
```

```python
trian_data = torchvision.datasets.CIFAR10('../dataset', train=True, transform=transforms.ToTensor())
dataloader = DataLoader(dataset=trian_data, batch_size=64, shuffle=False, drop_last=True, num_workers=0)
```

```python
class TestConv(nn.Module):
    def __init__(self):
        super(TestConv, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 6)
        self.relu1 = nn.ReLU()
        self.conv2 = nn.Conv2d(6, 3, 3)

    def forward(self, x):
        x = self.relu1(self.conv1(x))
        return self.conv2(x)
```

```python
test = TestConv()
step = 0

for data in dataloader:
    imgs, classes = data
    writer.add_images('Conv_Input', imgs, step)
    output_imgs = test(imgs)
    output_imgs = torch.reshape(output_imgs, (-1, 3, 25, 25))
    writer.add_images('Conv_Output', output_imgs, step)
    step += 1
```

```python
writer.close()
```
