```python
from PIL import Image
from torch.utils.tensorboard import SummaryWriter
from torchvision import transforms

writer = SummaryWriter("../logs")
```

## \_ \_call\_ \_

```python
class Person:
    def __call__(self, name):
        print('__call__:', name)

    def hello(self, name):
        print('hello:', name)

person = Person()
person('dougz')
person.hello('dougz')
```

## ToTensor

```python
images = ['../dataset/rainier.jpg', '../dataset/autumn_oak.jpg']
image_totensor = transforms.ToTensor()
for i in range(len(images)):
    image = Image.open(images[i])
    image_tensor = image_totensor(image)
    writer.add_image('ToTensor', image_tensor, i)
```

## Normalize

```python
image_normalize = transforms.Normalize((0.5, 0.5, 0.5), (1.5, 2.5, 3.5))
for i in range(len(images)):
    image = Image.open(images[i])
    image_tensor = image_totensor(image)
    image_normal = image_normalize(image_tensor)
    writer.add_image('Normalize', image_normal, i)
```

## Resize

```python
image_resize = transforms.Resize((128, 512))
for i in range(len(images)):
    image = Image.open(images[i])
    print(image.size)
    image_resized = image_resize(image)
    print(image_resized.size)
    image_tensor = image_totensor(image_resized)
    writer.add_image('Resize', image_tensor, i)
```

## Compose

```python
image_compose = transforms.Compose([
    image_resize,
    image_totensor
])
for i in range(len(images)):
    image = Image.open(images[i])
    image_tensor = image_compose(image)
    writer.add_image('Compose', image_tensor, i)
```

## RandomCrop

```python
image_randomCrop = transforms.RandomCrop(512)
image_compose = transforms.Compose([
    image_randomCrop,
    image_totensor
])

for i in range(len(images)):
    image = Image.open(images[i])
    for j in range(10):
        image_tensor = image_compose(image)
        writer.add_image('RandomCrop', image_tensor, i * 10 + j)
```

```python
writer.close()
```
