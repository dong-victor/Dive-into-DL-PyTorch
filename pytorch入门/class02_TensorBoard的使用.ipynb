```python
from torch.utils.tensorboard import SummaryWriter
from PIL import Image
import numpy as np
```

```python
writer = SummaryWriter('../logs')
```

```python
for i in range(0, 100):
    writer.add_scalar('test_x', i, 2 * i, 1)
```

```python
images = ['../dataset/0Q3A0032.JPG', '../dataset/rainier.jpg']
for i in range(0, len(images)):
    image = Image.open(images[i])
    writer.add_image('test_img', np.array(image), i, dataformats="HWC")
```

```python
writer.close()
```
