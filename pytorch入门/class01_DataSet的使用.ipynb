## DataSet DataLoader

```python
from torch.utils.data import Dataset
from PIL import Image
import os
```

```python
class MyData(Dataset):

    def __init__(self, root_dir, image_dir, label):
        self.root_dir = root_dir
        self.image_dir = image_dir
        self.label = label
        self.image_path = os.path.join(self.root_dir, self.image_dir)
        self.image_list = os.listdir(self.image_path)

    def __getitem__(self, index):
        image_name = self.image_list[index]
        image = Image.open(os.path.join(self.image_path, image_name))
        label = self.label
        return image, label

    def __len__(self):
        return len(self.image_list)
```

```python
root_dir = '../dataset/hymenoptera_data/train'
ants_dir = 'ants'
bees_dir = 'bees'
antsData = MyData(root_dir, ants_dir, ants_dir)
beesData = MyData(root_dir, bees_dir, bees_dir)
image, label = antsData[0]
print(image)
print(label)

train_data = antsData + beesData
print(len(train_data))
```
