```python
import torch
import torchvision
import torch.nn as nn
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter
from torchvision import transforms

writer = SummaryWriter('../logs')
```

```python
train_data = torchvision.datasets.CIFAR10('../dataset', train=True, transform=transforms.ToTensor())
dataloader = DataLoader(dataset=train_data, batch_size=64, shuffle=False, num_workers=0, drop_last=True)
```

```python
class TestSigmoid(nn.Module):
    def __init__(self):
        super(TestSigmoid, self).__init__()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        return self.sigmoid(x)
```

```python
test = TestSigmoid()
step = 0

for data in dataloader:
    imgs, _ = data
    writer.add_images('Sigmoid_Input', imgs, step)
    output = test(imgs)
    writer.add_images('Sigmoid_Output', output, step)
```

```python
writer.close()
```
